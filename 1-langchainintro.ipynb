#Streaming and Batch
Most models can stream their output content while it is beinf generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer responses. Calling stream() returns an iterator that yields outpt=ut chunks as they are produced. You can use a loop to process each chunk in real time.

```python
model.invoke("Write me a 500 words paragraph on AI.")
```

```python
for chunk in model.stream("Write me a 500 words paragraph on AI."):
    print(chunk.text, end="", flush=True)
```

```python
model.invoke("Why do parrots have colorful feathers?")
```

```python
for chunk in model.stream("Why do parrots have colorful feathers?"):
    print(chunk.text, end="", flush=True)
```

# Batching
Batching is collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel.

```python
responses = model.batch(
    [
        "Why do parrots have colorful feathers?",
        "How do airplanes fly?",
        "What is Quantum Computing?"
    ]
)
for response in responses:
    print(response)
```

```python
responses = model.batch(
    [
        "Why do parrots have colorful feathers?",
        "How do airplanes fly?",
        "What is Quantum Computing?"],
        config={
            'max_concurrency': 5, # Limit to parallel calls
        }
)
for response in responses:
    print(response)
```
